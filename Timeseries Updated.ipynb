```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ts_confirmed_global = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
ts_deaths_global = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'

confirmed = pd.read_csv(ts_confirmed_global)
deaths = pd.read_csv(ts_deaths_global)
```

```python
confirmed_cases = confirmed.groupby('Country/Region').sum()
deaths_cases = deaths.groupby('Country/Region').sum()
```

```python
plt.figure(figsize=(20,10))
plt.title('COVID-19 confirmed cases in the last 20 days in Brazil')
confirmed_cases.loc['Italy'][-20:].diff().dropna().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title('COVID-19 deaths cases in the last 20 days in Brazil')
deaths_cases.loc['Brazil'][-20:].diff().dropna().plot(kind='bar')
plt.show()
```

```python
lethality_rate = (deaths_cases.loc['Brazil'][-1] / confirmed_cases.loc['Brazil'][-1]) * 100
```
